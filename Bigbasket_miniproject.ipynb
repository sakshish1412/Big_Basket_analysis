# Step-6
Measuring discount on a certain item.

```python
def calc_discount(row):
    return ((row['market_price'] - row['sale_price']) / row['market_price']) * 100

df['discount'] = df.apply(calc_discount, axis=1)
```

# STEP-8
Find out the outliers from the dataset according to
the columns and fill them with the mean

```python
df.info()
```

```python
numeric_columns = ['sale_price', 'market_price', 'rating']

for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify and replace outliers with mean
    outliers = (df[col] < lower_bound) | (df[col] > upper_bound)
    df.loc[outliers, col] = df[col].mean()
```
